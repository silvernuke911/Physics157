```python
import numpy as np
import matplotlib.pyplot as plt

# Sample parameters
N = 1000  # Number of sample points
T = 1 / 1000.0  # Sample spacing
x = np.linspace(0.0, N*T, N)

y = [ix if ix < .5 else -ix+1 for ix in x]

plt.figure()
plt.plot(x,y, color = 'r')
plt.grid()
plt.show()

# Compute FFT
yf = np.fft.fft(y) / (N/2)
xf = np.fft.fftfreq(N, T)

# Compute RFFT
yrf = np.fft.rfft(y) / (N/2)
xrf = np.fft.rfftfreq(N, T)

# Plot FFT
plt.figure()
plt.plot(xf, np.abs(yf), label='FFT')
plt.title('FFT of the signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# Plot RFFT
plt.figure()
plt.plot(xrf, np.abs(yrf), label='RFFT', color='orange')
plt.title('RFFT of the signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.show()
```
