```python
# Load CSVs
df_hrzon = pd.read_csv(r'../sims_actual/nbody_01.csv')
df_2body = pd.read_csv(r'../sims_actual/2body_05.csv')
df_nbody = pd.read_csv(r'../sims_actual/nbody_05e.csv')

# Extract positions
x_h, y_h, z_h = df_hrzon['x'], df_hrzon['y'], df_hrzon['z']
x_2, y_2, z_2 = df_2body['x'], df_2body['y'], df_2body['z']

# Compute Euclidean distance between Horizons and 2-body
delta_2body = np.sqrt((x_h - x_2)**2 + (y_h - y_2)**2 + (z_h - z_2)**2)

# Plot
plt.figure(figsize=(5, 4), dpi=300)
plt.plot(df_hrzon['t'] / 365.25, delta_2body, label='Horizons - 2-body', color='red')
plt.xlabel('Time (Years)')
plt.ylabel('Distance Difference (AU)')
plt.title(r'\textbf{Deviation Between Horizons and 2-Body Simulation}')
plt.axhline(0.01, linestyle = '--', color = 'b', label = 'Deviation Threshold' )
plt.xlim(0,20)
plt.legend()
plt.tight_layout()
plt.show()
```
