```python
import numpy as np
import matplotlib.pyplot as plt

dx = 0.01
xv = np.arange(-4, 4, dx)
yv = np.arange(-2, 2, dx)
x, y = np.meshgrid(xv, yv)


U = x**2 + y**2 > -1
# --- Define sets as regions ---
# Example: A = inside circle centered at (-1,0) with radius 1.5
A = (x + 1)**2 + y**2 <= 1.5**2

# Example: B = inside circle centered at (1,0) with radius 1.5
B = (x - 1)**2 + y**2 <= 1.5**2

def union(A, B):
    return A | B 
def intersect (A, B):
    return A & B 
def complement (A):
    return ~A 
def subtract(A, B):
    return intersect(A, complement(B))

# --- Plot results ---
fig, axs = plt.subplots(4, 2, figsize=(5, 8))

axs[0,0].set_title(r"$A$")
axs[0,0].imshow(A, extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[0,1].set_title(r"$B$")
axs[0,1].imshow(B, extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[1,0].set_title(r"$A\cup B$")
axs[1,0].imshow(union(A,B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[1,1].set_title(r"$A\cap B$")
axs[1,1].imshow(intersect(A,B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[2,0].set_title(r"$A^c$")
axs[2,0].imshow(complement(A), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[2,1].set_title(r"$B^c$")
axs[2,1].imshow(complement(B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[3,0].set_title(r"$A- B$")
axs[3,0].imshow(subtract(A,B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[3,1].set_title(r"")
axs[3,1].imshow(U, extent=(-4,4,-2,2), origin='lower', cmap="Reds")

for ax in axs.flat:
    theta = np.linspace(0,2*np.pi, 200)
    ax.plot(1.5*np.cos(theta)-1, 1.5*np.sin(theta), linewidth = 1, color = 'k')
    ax.plot(1.5*np.cos(theta)+1, 1.5*np.sin(theta), linewidth = 1, color = 'k')
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()


# --- Plot results ---
fig, axs = plt.subplots(3, 2, figsize=(10, 8))

axs[0,0].set_title(r"$A\cup B$")
axs[0,0].imshow(union(A,B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[0,1].set_title(r"$(A\cup B)^c$")
axs[0,1].imshow(complement(union(A,B)), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[1,0].set_title("$A^c$")
axs[1,0].imshow(complement(A), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[1,1].set_title("$B^c$")
axs[1,1].imshow(complement(B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[2,0].set_title(r"$A^c \cap B^c$")
axs[2,0].imshow(intersect(complement(A), complement(B)), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[2,1].set_title(r"")
axs[2,1].imshow(U, extent=(-4,4,-2,2), origin='lower', cmap="Reds")

for ax in axs.flat:
    theta = np.linspace(0,2*np.pi, 200)
    ax.plot(1.5*np.cos(theta)-1, 1.5*np.sin(theta), linewidth = 1, color = 'k')
    ax.plot(1.5*np.cos(theta)+1, 1.5*np.sin(theta), linewidth = 1, color = 'k')
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()

# --- Plot results ---
fig, axs = plt.subplots(3, 2, figsize=(10, 8))

axs[0,0].set_title(r"$A\cap B$")
axs[0,0].imshow(intersect(A,B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[0,1].set_title(r"$(A\cap B)^c$")
axs[0,1].imshow(complement(intersect(A,B)), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[1,0].set_title("$A^c$")
axs[1,0].imshow(complement(A), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[1,1].set_title("$B^c$")
axs[1,1].imshow(complement(B), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[2,0].set_title(r"$A^c \cup B^c$")
axs[2,0].imshow(union(complement(A), complement(B)), extent=(-4,4,-2,2), origin='lower', cmap="Reds")

axs[2,1].set_title(r"")
axs[2,1].imshow(U, extent=(-4,4,-2,2), origin='lower', cmap="Reds")

for ax in axs.flat:
    theta = np.linspace(0,2*np.pi, 200)
    ax.plot(1.5*np.cos(theta)-1, 1.5*np.sin(theta), linewidth = 1, color = 'k')
    ax.plot(1.5*np.cos(theta)+1, 1.5*np.sin(theta), linewidth = 1, color = 'k')
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def transform_set(A, f, x_grid, y_grid):
    """
    Apply transformation f to set A defined on grid (x_grid, y_grid)
    
    Parameters:
    A: 2D boolean array (the set to transform)
    f: function that takes (x, y) and returns (u, v)
    x_grid, y_grid: 2D coordinate grids
    
    Returns:
    F_A: 2D boolean array (the transformed set)
    """
    # Create output array
    F_A = np.zeros_like(A, dtype=bool)
    
    # Get indices of points in the original set
    i_indices, j_indices = np.where(A)
    
    # Apply transformation to each point
    for i, j in zip(i_indices, j_indices):
        x_val = x_grid[i, j]
        y_val = y_grid[i, j]
        
        # Apply transformation
        u, v = f(x_val, y_val)
        
        # Find the closest grid point (nearest neighbor approximation)
        i_new = np.argmin(np.abs(y_grid[:, 0] - v))
        j_new = np.argmin(np.abs(x_grid[0, :] - u))
        
        # Mark this point as in the transformed set
        if (0 <= i_new < F_A.shape[0]) and (0 <= j_new < F_A.shape[1]):
            F_A[i_new, j_new] = True
    
    return F_A

# Example usage:
dx = 0.01  # Smaller step for better resolution
xv = np.arange(-2, 2, dx)
yv = np.arange(-2, 2, dx)
x, y = np.meshgrid(xv, yv)

# Original set: circle
A = ((x-0.5)**2 + y**2 <= 1)

# Define your arbitrary transformation
def f_example(x, y):
    # Example: (x, y) -> (x, y^2)
    return x, y**2

def f_rotation(x, y):
    # Example: 45 degree rotation
    theta = np.pi/4
    u = x*np.cos(theta) - y*np.sin(theta)
    v = x*np.sin(theta) + y*np.cos(theta)
    return u, v

def f_complex(x, y):
    # Example: more complex transformation
    return x + 0.5*y**2, 0.5*y + 0.5

# Apply transformation
F_A = transform_set(A, f_complex, x, y)

# Plot
fig, ax = plt.subplots(1, 2, figsize=(10, 6), dpi=300)
ax[0].imshow(A, extent=[-2, 2, -2, 2], origin='lower', cmap='Reds')
ax[1].imshow(F_A, extent=[-2, 2, -2, 2], origin='lower', cmap='Reds')

for i, axs in enumerate(ax):
    axs.set_xlabel("$x$")
    axs.set_ylabel("$y$")
ax[0].set_title("Original set $A$")
ax[1].set_title("Transformed set $f(A)$")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Vector utilities
# -----------------------------
def vproj(a, b):
    """Projection of vector a onto vector b."""
    a, b = np.array(a), np.array(b)
    if np.allclose(b, 0):
        return np.zeros_like(a)
    return b * np.dot(a, b) / np.dot(b, b)

def vrej(a, b):
    """Rejection of vector a from vector b (perpendicular component)."""
    return a - vproj(a, b)

def normalize(a, eps=1e-12):
    """Return normalized vector (safe against zero vector)."""
    a = np.array(a)
    norm = np.linalg.norm(a)
    if norm < eps:
        return np.zeros_like(a)
    return a / norm

# -----------------------------
# Define bases and vectors
# -----------------------------
origin = np.array([0, 0, 0])

# Standard Cartesian basis
xhat, yhat, zhat = np.eye(3)
basis1 = np.column_stack([xhat, yhat, zhat])
print("Basis 1:\n",basis1)

# Example vector
avec = np.array([0.3, 0.75, 0.86])

# Construct alternative basis
def euler_rotation(alpha, beta, gamma, order="zyx"):
    """Return rotation matrix for Euler angles (intrinsic rotations)."""
    ca, cb, cg = np.cos([alpha, beta, gamma])
    sa, sb, sg = np.sin([alpha, beta, gamma])
    
    Rz = np.array([[ca, -sa, 0],
                   [sa,  ca, 0],
                   [0,   0,  1]])
    
    Ry = np.array([[cb, 0, sb],
                   [0,  1, 0],
                   [-sb,0, cb]])
    
    Rx = np.array([[1, 0, 0],
                   [0, cg, -sg],
                   [0, sg,  cg]])
    
    if order == "zyx":
        return Rz @ Ry @ Rx
    elif order == "xyz":
        return Rx @ Ry @ Rz
    else:
        raise ValueError("Unsupported order")

# Example usage
alpha, beta, gamma = np.radians([-30, -30, 180])  # rotate 30° about z, 45° about y, 60° about x
basis2 = euler_rotation(alpha, beta, gamma)

# xprimehat = normalize([1, 10, -4])
# yprimehat = normalize(vrej([5, 4, 2], xprimehat))
# zprimehat = normalize(np.cross(xprimehat, yprimehat))
print("Basis 2:\n",basis2)

# Camera view direction (projection normal)
cam_norm = normalize([1, 1, 1])

print("a =",avec)
print("a'=",avec@basis2)

# -----------------------------
# Projection plane basis (u,v)
# -----------------------------
# Take projection of z-axis; fall back to y-axis if degenerate
v = normalize(vrej(zhat, cam_norm))
u = normalize(np.cross(v, cam_norm))

print("uv", u, v)
# -----------------------------
# Projection helper
# -----------------------------
def to_2d_coords(vec_3d):
    """Project 3D vector onto the (u,v)-plane orthogonal to cam_norm."""
    return np.array([np.dot(vec_3d, u), np.dot(vec_3d, v)])

# -----------------------------
# Plot helpers
# -----------------------------
def plot_vector(start, end, color, label, scale=1.0, alpha=1.0, linestyle='-'):
    """Draw a single vector arrow in 2D."""
    dx, dy = (end * scale)
    ax.arrow(start[0], start[1], dx, dy,
             head_width=0.03, head_length=0.05,
             fc=color, ec=color, alpha=alpha,
             linestyle=linestyle, length_includes_head=True)
    ax.text(start[0] + dx * 1.1, start[1] + dy * 1.1,
            label, color=color, fontsize=12)

def plot_basis(basis, colors=('r', 'g', 'b'), labels=(r'\hat{x}', r'\hat{y}', r'\hat{z}'),
               origin_point=np.zeros(3), linestyle='-', alpha=1.0):
    """Plot a 3D basis projected into 2D."""
    for vec, color, label in zip(basis.T, colors, labels):
        v2d = to_2d_coords(vrej(vec, cam_norm))
        lbl = rf'$\mathbf{{{label}}}$'
        plot_vector(to_2d_coords(vrej(origin_point, cam_norm)),
                    v2d, color, lbl, linestyle=linestyle, alpha=alpha)

def plot_dimvec(vec, basis, color='k', origin_point=np.zeros(3),
                linestyle='--', alpha=0.8, lw=0.8):
    """Plot decomposition of a vector into given basis components."""
    basis = basis.T
    p = [origin_point]
    p.append(p[-1] + vec[0] * basis[0])
    p.append(p[-1] + vec[1] * basis[1])
    p.append(p[-1] + vec[2] * basis[2])
    proj_pts = [to_2d_coords(vrej(pi, cam_norm)) for pi in p]
    for a, b in zip(proj_pts[:-1], proj_pts[1:]):
        ax.plot([a[0], b[0]], [a[1], b[1]], color=color,
                linestyle=linestyle, alpha=alpha, lw=lw)

# -----------------------------
# Projection of key vectors
# -----------------------------
origin_2d = to_2d_coords(origin)
avec_proj = vrej(avec, cam_norm)
avec_2d = to_2d_coords(avec_proj)

# -----------------------------
# Create plot
# -----------------------------
fig, ax = plt.subplots(figsize=(4, 4), dpi=300)

# Plot bases
plot_basis(basis1)
plot_basis(basis2, labels=(r"\hat{x}'", r"\hat{y}'", r"\hat{z}'"), alpha=0.5)

# Plot vector a and its decompositions
plot_vector(origin_2d, avec_2d, 'k', r'$\mathbf{v}$')
plot_dimvec(avec, basis1)
plot_dimvec(avec @ basis2, basis2, alpha=0.5)
#plot_dimvec(basis2.T[0],basis1, alpha = 0.25, color='r')

# Style
ax.set_aspect('equal')
extent = max(np.linalg.norm(to_2d_coords(vrej(v, cam_norm)))
             for v in [xhat, yhat, zhat, avec])
ax.set_xlim(-extent - 0.2, extent + 0.2)
ax.set_ylim(-extent - 0.2, extent + 0.2)
ax.set_xlabel(r'$\mathbf{u}$ axis')
ax.set_ylabel(r'$\mathbf{v}$ axis')
ax.axis(False)
plt.tight_layout()
plt.show()
```
