```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

def science_plot(fontsize = 9):
    import scienceplots
    plt.style.use(['science','grid','notebook'])
    plt.rcParams.update({
        'font.size'       : fontsize,    # General font size
        'axes.titlesize'  : fontsize,    # Font size of the axes title
        'axes.labelsize'  : fontsize,    # Font size of the axes labels
        'xtick.labelsize' : fontsize,    # Font size of the x-axis tick labels
        'ytick.labelsize' : fontsize,    # Font size of the y-axis tick labels
        'legend.fontsize' : fontsize,    # Font size of the legend
        'figure.titlesize': fontsize,    # Font size of the figure title
        'legend.fancybox' : False,       # Disable the fancy box for legend
        'legend.edgecolor': 'k',         # Set legend border color to black
        'text.usetex'     : True,        # Use LaTeX for text rendering
        'font.family'     : 'serif'      # Set font family to serif
    })
science_plot()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("Food_Coloring_Chloro_CSV.csv")

# Plot each raw spectrum vs its wavelength
plt.figure(figsize=(6, 3), dpi = 300)

plt.plot(df["Blue Raw:Wavelength(nm)"], df["Blue Raw:Raw"], label="Blue", color="blue")
plt.plot(df["Yellow Raw:Wavelength(nm)"], df["Yellow Raw:Raw"], label="Yellow", color="gold")
plt.plot(df["Green Raw:Wavelength(nm)"], df["Green Raw:Raw"], label="Green", color="green")
plt.plot(df["Chlorophyll Raw:Wavelength(nm)"], df["Chlorophyll Raw:Raw"], label="Chlorophyll", color="lightgreen")

# Plot settings
# plt.title("Raw Spectra of Food Coloring and Chlorophyll")
plt.xlim(380,900)
plt.ylim(0, 30000)
plt.xlabel("Wavelength (nm)")
plt.ylabel("Raw Intensity")
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("Olive_Oil_With_Chlorophyll.csv")

# Plot all raw spectra
plt.figure(figsize=(6,3), dpi=300)

plt.plot(df["Olive Oil A Extra Light:Wavelength(nm)"], df["Olive Oil A Extra Light:Raw"], label="Olive Oil Extra Light", color="orange")
plt.plot(df["Olive Oil B Classic:Wavelength(nm)"], df["Olive Oil B Classic:Raw"], label="Olive Oil Classic", color="olive")
plt.plot(df["Olive Oil C Pomace:Wavelength(nm)"], df["Olive Oil C Pomace:Raw"], label="Olive Oil Pomace", color="brown")
plt.plot(df["Olive Oil D Extra Virgin:Wavelength(nm)"], df["Olive Oil D Extra Virgin:Raw"], label="Olive Oil Extra Virgin", color="green")
plt.plot(df["Chlorophyll Sample:Wavelength(nm)"], df["Chlorophyll Sample:Raw"], label="Chlorophyll Sample", color="lightgreen")
# Plot settings
# plt.title("Raw Spectra of Olive Oils and Chlorophyll Sample")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Raw Intensity")
plt.legend(loc="upper right")
plt.xlim(380,900)
plt.ylim(0, 30000)
plt.tight_layout()

# Show the plot
plt.show()
```
